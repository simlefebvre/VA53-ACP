```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('resultats.csv', sep=',')
```

```python
plot = df.boxplot(column=['test_acc'], by='nbCouche', grid=False,)
fig = plot.get_figure()
fig.suptitle('')

plt.title("Répartition des résultats de \nclassification en fonction du nombre de couche")
plt.ylabel("Pourcentage de réussite")
plt.xlabel("nombre de couche")
plt.savefig("fig/nbCouche.jpg")
plt.show()
```

```python
plot = df.boxplot(column=['test_acc'], by='nbNeurone', grid=False,)
fig = plot.get_figure()
fig.suptitle('')

plt.title("Répartition des résultats de classification \nen fonction du nombre de neurone par couche")
plt.ylabel("Pourcentage de réussite")
plt.xlabel("nombre de neurone par couche")
plt.savefig("fig/nbNeurone.jpg")
plt.show()
```

```python
plot = df.boxplot(column=['test_acc'], by='tailleImage', grid=False,)
fig = plot.get_figure()
fig.suptitle('')

plt.title("Répartition des résultats de classification \nen fonction de la taille de de l'image ")
plt.ylabel("Pourcentage de réussite")
plt.xlabel("Taille de l'image en pixel")
plt.savefig("fig/tailleImage.jpg")
plt.show()
```

```python
dfACP = pd.read_csv('resultatACP.csv', sep=',')
```

```python
plot = dfACP.plot(x='i', y='dureeCalculModel', kind='scatter')

fig = plot.get_figure()

plt.title("Temps de calcul du modèle \nen fonction du nombre de composante")
plt.xlabel("Nombre de composante")
plt.ylabel("Temps de calcul du modèle en secondes")

plt.show()
```
